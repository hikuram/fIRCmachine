```python
# Helper function for file upload
def upload_as(new_name: str):
    import os
    from google.colab import files
    print(f"Please upload the file for: {new_name}")
    uploaded = files.upload()
    if uploaded:
        uploaded_name = list(uploaded.keys())[0]
        os.rename(uploaded_name, new_name)
        print(f"-> Successfully saved as {new_name}\n")
```

```python
# Upload your input structures
upload_as("react.xyz")
upload_as("prod.xyz")
```
