```python
# Setup matplotlib
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import subprocess
import logging

# 1. Mute font_manager warnings (e.g., missing 'cursive' font)
logging.getLogger('matplotlib.font_manager').setLevel(logging.ERROR)

# 2. Silently install Microsoft core fonts (including Arial)
install_cmd = """
echo "ttf-mscorefonts-installer msttcorefonts/accepted-mscorefonts-eula select true" | debconf-set-selections && \
apt-get update -qq && \
apt-get install -y -qq ttf-mscorefonts-installer
"""
subprocess.run(install_cmd, shell=True, stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)

# 3. Register Arial directly to avoid cache rebuilding
font_path = '/usr/share/fonts/truetype/msttcorefonts/Arial.ttf'
fm.fontManager.addfont(font_path)

# 4. Set Arial as the global default font
plt.rcParams['font.family'] = 'Arial'

print("✅ Setup complete: Arial font installed and applied silently.")
```

```python
# Helper function for file upload
def upload_as(new_name: str):
    import os
    from google.colab import files
    print(f"Please upload the file for: {new_name}")
    uploaded = files.upload()
    if uploaded:
        uploaded_name = list(uploaded.keys())[0]
        os.rename(uploaded_name, new_name)
        print(f"-> Successfully saved as {new_name}\n")
```

```python
# Upload your input structures
upload_as("react.xyz")
upload_as("prod.xyz")
```
